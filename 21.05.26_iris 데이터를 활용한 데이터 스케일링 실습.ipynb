```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import tqdm
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import export_graphviz
```

```python
iris = load_iris()
iris
```

```python
iris_data = iris['data']
iris_data
```

```python
# 데이터프레임으로 만들기
iris_df = pd.DataFrame (iris_data, columns=iris['feature_names'])
iris_df
```

### Standard Scaler 적용 

```python
# 스탠다드 스케일러 임포트
from sklearn.preprocessing import StandardScaler
```

```python
# 임포트 한 스탠다드 스케일러를 불러와서 변수에 담아 줌
scaler = StandardScaler()

# iris_df로 스케일러를 학습
scaler.fit(iris_df)

# 학습된 스케일러의 데이터 값을 실제로 변환시켜줌
iris_SS = scaler.transform(iris_df)
```

```python
#  array 형태로 변환되어서 출력됨
iris_SS 
```

```python
# iris_SS를 데이터프레임으로 만들기!
iris_df_SS = pd.DataFrame (iris_SS, columns= iris["feature_names"])
```

```python
iris_df_SS
```

```python
print('원본 데이터 특성별 평균')
print(iris_df.mean())
print('-'*30)
print('원본 데이터 특성별 분산')
print(iris_df.var())
```

```python
print('스케일링 데이터 특성별 평균')
print(iris_df_SS.mean())
print('-'*30)
print('스케일링 데이터 특성별 분산')
print(iris_df_SS.var())
```

- 평균은 0에 가깝게, 분산은 1에 가깝게 변환되는 것을 확인할 수 있음
- 스탠다드 스케일러는 데이터의 중심과 퍼짐 정도를 제한하여 데이터의 분포를 고르게 만들어 줌

### MinMax Scaler
- 0 ~ 1 사이의 값으로 변환 (-가 있다면 -1~1사이로 변환)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler1 = MinMaxScaler()
scaler1.fit(iris_df)
iris_MM = scaler1.transform (iris_df)
iris_MM
```

```python
iris_df_MM = pd.DataFrame(iris_MM, columns=iris['feature_names'])
```

```python
iris_df_MM
```

```python
print('원본 데이터 특성별 최소값')
print(iris_df.min())
print('-'*30)
print('원본 데이터 특성별 최대값')
print(iris_df.max())
```

```python
print('스케일링 데이터 특성별 최소값')
print(iris_df_MM.min())
print('-'*30)
print('스케일링 데이터 특성별 최대값')
print(iris_df_MM.max())
```

- 만약 train, test 데이터가 나뉘어져 있다면 train 데이터로 학습한 scaler 모델을 그대로 이용하여 test 데이터에서 transform 해줘야함

- test 데이터로 다시 스케일링 학습을 해버리면 train 데이터와 다른 기준을 가지고 학습이 되게 때문에 train 데이터로 fit한 모델을 그대로 test에도 사용해줘야함 (MinMax를 예로 들면 train데이터의 최대, 최소값과 test데이터의 최대, 최소값이 다를 수 있기 때문에 각각 따로 적용했을 경우 다른 결과가 나올 수 있음)
